```python
import pandas as pd

df = pd.read_csv(
    "../data/raw/complaints.csv",
    low_memory=False,
    nrows=100000
)

df.shape
```

```python
df["word_count"] = (
    df["Consumer complaint narrative"]
    .fillna("")
    .str.split()
    .str.len()
)
```

```python
products = [
    "Credit card",
    "Personal loan",
    "Savings account",
    "Money transfer"
]

filtered = df[df["Product"].isin(products)]

filtered = filtered[
    filtered["Consumer complaint narrative"].notna()
]
```

```python
filtered.shape
```

```python
filtered.to_csv(
    "../data/processed/filtered_complaints.csv",
    index=False
)
```
